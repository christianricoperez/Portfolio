### Test ANOVA

El Vecindario tiene alguna repercusin en el precio?

* Hiptesis nula (H0): No hay diferencias significativas en los precios de venta entre los diferentes vecindarios
* Hiptesis alternativa (H1): Existen diferencias significativas en los precios de venta entre al menos dos vecindarios

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('PrecioVenta ~ Vecindario', data=house_df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

p-value < 0.05, por lo tanto se rechaza la hiptesis nula. Significa que hay evidencia suficiente para decir que al menos dos vecindarios tienen precios de venta diferentes.

* Identificacin del Patrn: Las categoras analizadas muestran una estructura inherentemente ordenada. Un caso claro es la caracterstica "piscina", donde una vivienda con piscina establece una calidad o jerarqua superior respecto a una sin ella.

* Decisin de Modelado: Para capturar este orden y representarlo numricamente, se implementar una tcnica de codificacin: transformaremos estas variables ordinales (o con orden implcito) asignando a cada categora el valor de la media del precio de venta observada en dicha categora.

```python
def encode(frame, feature):
    ordering = pd.DataFrame()
    #Guardamos los valores nicos de cada variable cualitativa
    ordering['ValoresUnicos'] = frame[feature].unique()
    #Los ndices de nuestro df sern estas categoras
    ordering.index = ordering.ValoresUnicos
    #Hallamos la media del precio de venta por cada valor nico de la variable categrica
    ordering['PrecioVentaMedia'] = frame[[feature, 'PrecioVenta']].groupby(feature).mean()['PrecioVenta']
    #Ordenamos nuestro df de acuerdo a la media del precio de venta
    ordering = ordering.sort_values('PrecioVentaMedia')
    #Asignamos un valor de orden a cada valor nico
    ordering['ordering'] = range(1, ordering.shape[0]+1)
    ordering = ordering['ordering'].to_dict()

    #Creamos nuestros campos encodeados en nuestro dataframe
    for cat, o in ordering.items():
        frame.loc[frame[feature] == cat, feature+'_E'] = o

#Guardamos los nombres de columnas encodeadas en una lista
qual_encoded = []
for q in qualitative:
    encode(house_df, q)
    qual_encoded.append(q+'_E')
#Mostramos los campos encodeados
print(qual_encoded)
```

### Multicolinealidad

* La fuerte relacin entre las variables que usamos para predecir (variables predictoras) es un problema.

* Esta correlacin excesiva tiene el potencial de inestabilizar nuestro modelo estadstico.

* Adems, puede resultar en interpretaciones incorrectas o engaosas sobre la influencia real de cada variable en el resultado principal.

**Cmo detectar la multicolinealidad?**

Anlisis VIF(Factor de influencia de varianza): Evala la medida en que cada variable independiente en un modelo est linealmente relacionada con las otras variables independientes.
Un VIF > 10 indica alta multicolinealidad.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
# quantitative + qual_encoded
vif_data["feature"] = quantitative
vif_data["VIF"] = [variance_inflation_factor(house_df[quantitative].values, i)
                          for i in range(len(quantitative))]
vif_data
```

```python
columnas_seleccionadas = vif_data[vif_data.VIF <= 100].feature.tolist()
columnas_seleccionadas
```

```python
from scipy import stats
import numpy as np

def outliers_col(df, columnas):
  for col in columnas:
    #Solo son outliers los que sobrepasan el z score de 3
    n_outliers = len(df[np.abs(stats.zscore(df[col])) > 3])
    print("{} | {}".format(df[col].name, n_outliers))
outliers_col(house_df, quantitative + ["PrecioVenta"])
```

# 5.Comprobacin de supuestos.

Hay que comprobar cuatro suposiciones fundamentales:

- Normalidad: Los datos deben parecerse a una distribucin normal. Es importante porque varias pruebas estadsticas se basan en esta suposicin. Si resolvemos la normalidad evitamos otros problemas, como la homocedasticidad.

- Homocedasticidad: Suposicin de que las variables dependientes tienen el mismo nivel de varianza en todo el rango de las variables predictoras. La homocedasticidad es deseable porque queremos que el trmino de error sea el mismo en todos los valores de las variables independientes.

- Linealidad: La forma ms comn de evaluar la linealidad es examinar los **diagramas de dispersin** y buscar patrones lineales. Si los patrones no son lineales, valdra la pena explorar las transformaciones de datos.

- Ausencia de errores correlacionados - Esto ocurre a menudo en series temporales, donde algunos patrones estn relacionados en el tiempo.

### Normalidad

El objetivo es estudiar la variable 'PrecioVenta' de forma fcil, comprobando:

- Histograma: Curtosis y asimetra.
- Grfica de probabilidad normal: La distribucin de los datos debe ajustarse a la diagonal que representa la distribucin normal.
-Test de normalidad: Dos pruebas comunes son la Prueba de Shapiro-Wilk y la Prueba de Kolmogorov-Smirnov.

### Test de normalidad  

Los Test de normalidad se utilizan para verificar si una muestra de datos sigue una distribucin normal. Dos de los tests ms comunes son:
* Test de Shapiro-Wilk:

Es una prueba estadstica utilizada para evaluar si una muestra de datos proviene de una distribucin normal. Es adecuado para muestras de tamao moderado($n < 30$). Se plantean las siguientes hiptesis:
  - Hiptesis nula ($H_0$): La muestra sigue una distribucin normal.
  - Hiptesis alternativa ($H_1$): La muestra no sigue una distribucin normal.
* Test de Kolmogorov-Smirnov:

Es una prueba no paramtrica que determina la **bondad de ajuste** de dos distribuciones de probabilidad entre s. Es aplicable para muestras grandes.  Para una distribucin normal se plantean las siguientes hiptesis:
  - Hiptesis nula ($H_0$): La muestra sigue una distribucin normal.
  - Hiptesis alternativa ($H_1$): La muestra no sigue una distribucin normal.


Los precios de venta se ajustan a una distribucin normal?

```python
# Histograma
sns.distplot(house_df['PrecioVenta'], fit = norm)
plt.plot()
```

```python
from scipy import stats
import numpy as np

stat_ks, p_value = sm.stats.diagnostic.kstest_normal(house_df["PrecioVenta"], dist = 'norm')
stat_ks, p_value
```

p-value = 0.0009999999999998899 < 0.05, por lo tanto los precios de venta de las viviendas no siguen una distribucin normal.
